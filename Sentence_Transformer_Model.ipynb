```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
categories_df = pd.read_excel('/content/drive/MyDrive/Google_Categories.xlsx')
categories = categories_df['CategoryDisplayName'].tolist()
```

```python
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

```python
category_embeddings = model.encode(categories)
```

```python
def find_related_categories(keyword):
    keyword_embedding = model.encode([keyword])[0]
    similarities = cosine_similarity([keyword_embedding], category_embeddings)[0]
    scores={}
    for i, similarity in enumerate(similarities):
          scores[categories[i]] = similarity

    sorted_scores= {k: v for k, v in reversed(sorted(scores.items(), key=lambda item: item[1]))}

    return sorted_scores
```

```python
keyword = input("Enter a keyword: ")
related_categories = find_related_categories(keyword)
print(f"Categories related to '{keyword}':")
for category, score in related_categories.items():
    if score >= 0.5:
      print(f"- {category} {score}")
    else:
      break
```
